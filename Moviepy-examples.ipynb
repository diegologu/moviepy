# Some moviepy examples 

```python
# run me first
from IPython.display import Video, Audio

def cat_run_show(scriptname, infname, outfname="output/tmp.mp4"):
    %pycat {scriptname}
    %run {scriptname} {infname} {outfname}
    return Video(outfname, embed=True, width=680)
```

```python
# Original video 1
Video("input/example.mp4", embed=True, width=680)
```

```python
# Original video 2
Video("input/example2.mp4", embed=True, width=680)
```
